Dataset: Greenland Ice Discharge (PROMICE) since 1986

There are two main datatypes:

1) 3 NetCDF files ( GIS.nc, region.nc, gate.nc & sector.nc)

  -  Other than GIS.nc, we were able to open all .nc files using qgis. Following error was shown when we attempted to open GIS.nc using qgis. 
  Error: Unsupported data source. GIS.nc has no band and is invalid as raster layer.
  
  -  CRS info is not available 
  
  -  Yes, open and interopearble to some extent

2) CSV

   - Yes, viewable in qgis.
   
   - CRS info is not available 
   
   - Yes, the data is open and interoperable.
